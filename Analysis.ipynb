```python
housing = pd.read_excel('clean_housing.xlsx')
education = pd.read_excel('clean_education.xlsx')
```

```python
#housing.columns

grouped_housing = housing.groupby('county_fips').mean(numeric_only=True)
grouped_education = education.groupby('county_fips').mean(numeric_only = True)
#grouped_education
grouped_housing
```

```python
grouped_housing['percent'] = (grouped_housing['percent'] / 100)
grouped_education['estimate'] = (grouped_education['estimate'] / 100)
#print(grouped_education)
#print(grouped_housing)
```

```python
grouped_housing['percent']
table = np.array([grouped_housing['percent'], grouped_education['estimate']])
print(table)

stat, pval, degrees, freq_array = stats.chi2_contingency(table)
#stat, pval, degrees, freq_array
stats.chi2_contingency(table)
```
